```python


Tea_bags_box=150
Milk_Powder=100
Tea_Masala=60
Tea_Powder=120
Coffee_Powder=80
cost_Tea_bags_box=90
cost_Milk_Powder=60
cost_Tea_Masala=15
cost_Tea_Powder=70
cost_Coffee_Powder=65
def add_items_bag():
    print("%10s %10s %10s %10s" %('Sr.no.','Items','Quantity','Price/Unit'))
    print("%10s %10s %10d %10.2f" %(1,'Tea_bags_box',Tea_bags_box,cost_Tea_bags_box))
    print("%10s %10s %10d %10.2f" %(2,'Milk_Powder',Milk_Powder,cost_Milk_Powder))
    print("%10s %10s %10d %10.2f" %(3,'Tea_Masala',Tea_Masala,cost_Tea_Masala))
    print("%10s %10s %10d %10.2f" %(4,'Tea_Powder',Tea_Powder,cost_Tea_Powder))
    print("%10s %10s %10d %10.2f" %(5,'Coffee_Powder',Coffee_Powder,cost_Coffee_Powder))
    print("Inventory Details")
    qty_Tea_bags_box=int(input("Enter the Tea Bags quantity: "))
    qty_Milk_Powder=int(input("Enter the Milk Powder quantity: "))

    qty_Tea_Masala=int(input("Enter the Tea Masala quantity: "))
    qty_Tea_Powder=int(input("Enter the Tea Powder quantity: "))
    qty_Coffee_Powder=int(input("Enter the Coffee Powder quantity: "))
    return qty_Tea_bags_box,qty_Milk_Powder,qty_Tea_Masala,qty_Tea_Powder,qty_Coffee_Powder

def total_cost():
    total_cost1=qty_Tea_bags_box*Tea_bags_box+qty_Milk_Powder*Milk_Powder+qty_Tea_Masala*Tea_Masala+qty_Tea_Powder*Tea_Powder+qty_Coffee_Powder*Coffee_Powder
    return total_cost1

def Cus_details():
    name=input("Customer Name: ")
    Address=input("Customer Address: ")
    distance=int(input("Enter the Distance from shop: "))
    return name,Address,distance

def delivery_charge(distance):
    if distance<7:
        delivery_charge=0
    else:
        delivery_charge=100
    
    return delivery_charge

def final_bill(total_cost1,delivery_charge):
    final_bill_amount=total_cost1 + delivery_charge
    return final_bill_amount

def generate_bill():
    
    print("Your Final Bill")
    print("%10s %10s %10s %10s %10s" %('Sr.no.','Items','Quantity','Price/Unit','Total Cost'))
    print("%10s %10s %9d %10.2f %10.2f" %(1,'Tea_bags_box',qty_Tea_bags_box,cost_Tea_bags_box,qty_Tea_bags_box*cost_Tea_bags_box))
    print("%10s %10s %9d %10.2f %10.2f" %(2,'Milk_Powder',qty_Milk_Powder,cost_Milk_Powder,qty_Milk_Powder*cost_Milk_Powder))
    print("%10s %10s %11d %10.2f %10.2f" %(3,'Tea_Masala',qty_Tea_Masala,cost_Tea_Masala,qty_Tea_Masala*cost_Tea_Masala))
    print("%10s %10s %11d %10.2f %10.2f" %(4,'Tea_Powder',qty_Tea_Powder,cost_Tea_Powder,qty_Tea_Powder*cost_Tea_Powder))
    print("%10s %10s %9d %10.2f %10.2f" %(5,'Coffee_Powder',qty_Coffee_Powder,cost_Coffee_Powder,qty_Coffee_Powder*cost_Coffee_Powder))
    
    print("The Total Cost of items:",total_cost())
    print("Final Bill",final_bill(total_cost1,delivery_charge))


while True:
    qty_Tea_bags_box,qty_Milk_Powder,qty_Tea_Masala,qty_Tea_Powder,qty_Coffee_Powder=add_items_bag()
    total_cost1=total_cost()
    Tea_bags_box=Tea_bags_box-qty_Tea_bags_box
    Milk_Powder=Milk_Powder-qty_Milk_Powder
    Tea_Masala=Tea_Masala-qty_Tea_Masala
    Tea_Powder=Tea_Powder-qty_Tea_Powder
    Coffee_Powder=Coffee_Powder-qty_Coffee_Powder
    name,Address,distance=Cus_details()
    delivery_charge=delivery_charge(distance)
    final_bill_amount=final_bill(total_cost1,delivery_charge)
    generate_bill()
    choice=input("Enter the choice yes/No")
    if choice== "Yes":
        continue
    else:
        print("Thanks for Coming,Visit Again")
    
        break














```
